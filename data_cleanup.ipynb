```python
import pandas as pd
```

```python
#read the CSV files
mvps = pd.read_csv("mvps.csv")
```

```python
#clean up the data to get rid of the extra columns
#make sure to only select certain columns in the data frame
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
#Clean up the players data
players = pd.read_csv("players.csv")
```

```python
players
```

```python
#Clean up this data frame and combine with the mvp data
#We can see the amount of votes players have
del players["Rk"]
```

```python
players.head()
```

```python
#Fix up the player names and merge the players df with mvp df
players["Player"].head(50)
```

```python
#replace the * behind some players to merge 
players["Player"] = players["Player"].str.replace("*", "", regex = False)
```

```python
players.head(50)
```

```python
players.head(20)
```

```python
#Every player should only have one row, each players teams shouldn't be counted
players.groupby(["Player", "Year"])
```

```python
#create a function to make the rows single row
def single_row(df):
    if df.shape[0] ==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        #replace the TOT with player's last played team
        row["Tm"] = df.iloc[-1,:]["Tm"]
        
        return row 
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
#Outer merge because the MVP data only has data has people who have won.
#Outer merge -> if can't find data in one set keep the data
combined = players.merge(mvps, how = "outer", on = ["Player", "Year"])
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
#need to clean the team data 
#combine the team data 
teams = pd.read_csv("teams.csv")
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
lines
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head(5)
```

```python
#combine into a new data frame
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```
